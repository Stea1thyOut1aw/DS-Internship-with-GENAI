## Defanging an IP Address

```python
address = input("enter a ip address")

def defangIPaddr(address):
    output = ""
    for i in address:
        if i == ".":
            output += "[.]"
        else:
            output += i
    return output

defangIPaddr(address)
```

## Find Numbers with Even Number of Digits

```python
l = list(map(int,input("enter numbers: ").split()))

def count_even_digit_numbers(l):
  count_even = 0
  for i in l:
    count = 0
    while(i>0):
      count += 1
      i = i/10
    if count%2 == 0:
      count_even += 1
  return count_even

count_even_digit_numbers(l)
```

## Number of Good Pairs

```python
nums = list(map(int,input("enter numbers:").split()))

def numIdenticalPairs(nums):
    count = 0
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] == nums[j]:
                count += 1
    return count

numIdenticalPairs(nums)
```

## How Many Numbers Are Smaller Than the Current Number

```python
s = list(map(int,input("enter numbers:").split()))

def smallerNumbersThanCurrent(s):
  a = {}
  for index,value in enumerate(sorted(s)):
    if value not in a:
      a[value] = index


  ans = [a[value] for value in s]
  return ans

smallerNumbersThanCurrent(s)
```

## Subtract the Product and Sum of Digits of an Integer

```python
n = int(input("enter a number:"))

def subtractProductAndSum(n):
    sum_digits = 0
    product_digits = 1

    while n > 0:
        digit = n % 10
        sum_digits += digit
        product_digits *= digit
        n //= 10

    return product_digits - sum_digits

subtractProductAndSum(n)
```

## XOR Operation in an Array

```python
n = int(input("enter a number:"))
start = int(input("enter the start number:"))

def create_array(n,start):
  arr = []
  for i in range(n):
    arr = arr + [start + 2*i]
  return arr

def xor(n,start):
  arr = create_array(n,start)
  xor = 0
  for i in arr:
    xor = xor ^ i
  return xor

xor(n,start)
```
